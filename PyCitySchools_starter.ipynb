```python
# Dependencies and Setup
import pandas as pd
import os
#import numpy 

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools
TOTAL_SCHOOLS=school_data_complete['school_name'].nunique()
print(TOTAL_SCHOOLS)

#Calculate the total number of students
TOTAL_STUDENTS=school_data_complete['student_name'].count()
print(TOTAL_STUDENTS)

#Calculate the total budget
TOTAL_BUDGET=school_data_complete["budget"].sum()
print(TOTAL_BUDGET)

#Calculate the average math score
AVG_MATHSCORE=school_data_complete["math_score"].mean()
print(AVG_MATHSCORE)

#Calculate the average reading score
AVG_READINGSCORE=school_data_complete["reading_score"].mean()
print(AVG_READINGSCORE)

#Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
overall_average_score=(AVG_MATHSCORE+AVG_READINGSCORE)/2
print(overall_average_score)

#Calculate the percentage of students with a passing math score (70 or greater)
percent_math=school_data_complete["math_score"].ge(70).mean()*100
print(percent_math)

#Calculate the percentage of students with a passing reading score (70 or greater)
percent_reading=school_data_complete["reading_score"].ge(70).mean()*100
print(percent_reading)

#Create a dataframe to hold the above results
df = pd.DataFrame({'TOTAL_SCHOOLS': [TOTAL_SCHOOLS],'TOTAL_STUDENTS':[TOTAL_STUDENTS],'TOTAL_BUDGET':[TOTAL_BUDGET],'AVG_MATHSCORE':[AVG_MATHSCORE],'AVG_READINGSCORE':[AVG_READINGSCORE],'overall_average_score':[overall_average_score],'percent_math':[percent_math],'percent_reading':[percent_reading]})

df.head()
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
#schooltype
SCHOOL_NAME_TYPE = school_data.set_index(["school_name"])["type"]
print(SCHOOL_NAME_TYPE)


#total students per school
total_students_per_school = school_data_complete["school_name"].value_counts()
print(total_students_per_school)


#total school budget
budget_per_school = school_data_complete.groupby(["school_name"]).mean()["budget"]
budget_per_school
```

```python
#Per Student Budget 
budget_per_student = budget_per_school/total_students_per_school
budget_per_student
```

```python
#Average Math Score per school
avg_math_scores = round(school_data_complete.groupby(["school_name"]).mean()["math_score"],2)
avg_math_scores
```

```python
#Average Reading Score per school
avg_reading_scores = round(school_data_complete.groupby(["school_name"]).mean()["reading_score"],2)
avg_reading_scores
```

```python
#% Passing Math per school

# total students passing math from the combined dataset
#students_passing_math= school_data_complete["math_score"].ge(70)
students_passing_math = school_data_complete[(school_data_complete["math_score"]>70)]

#total students passing math per school by using groupby
students_passing_math_per_school = students_passing_math.groupby(["school_name"]).count()["student_name"]

#percentage of students passing math per school
percentage_passing_math_per_school = round((students_passing_math_per_school/total_students_per_school)*100,2)
percentage_passing_math_per_school


```

```python
#% Passing Reading per school

# total students passing reading from the combined dataset
students_passing_reading = school_data_complete[school_data_complete["reading_score"]>70]

#total students passing reading per school by using groupby
students_passing_reading_per_school = students_passing_reading.groupby(["school_name"]).count()["student_name"]

#percentage of students passing reading per school
percentage_passing_reading_per_school = round((students_passing_reading_per_school/total_students_per_school)*100,2)
percentage_passing_reading_per_school
```

```python
#Overall Passing Rate (Average of the above two)

overall_passing_rate=(percentage_passing_math_per_school+percentage_passing_reading_per_school)/2
overall_passing_rate
```

```python
#Create a dataframe to hold the above results for school summary

df_school_summary = pd.DataFrame({"School Type":SCHOOL_NAME_TYPE,
                                 "Total Student per School":total_students_per_school,
                                 "Total School Budget":budget_per_school,
                                  "Budget per student":budget_per_student,
                                 "Average Math Score per School":avg_math_scores,
                                 "Average Reading Score per School":avg_reading_scores,
                                 "% Passing Math per School":percentage_passing_math_per_school,
                                 "% Passing Reading per School":percentage_passing_reading_per_school,
                                 "Overall Passing rate":overall_passing_rate})
df_school_summary
```

```python
#Top Performing Schools by Passing Rate
top_performing_schools = df_school_summary.sort_values(["Overall Passing rate"],ascending = False).head(5)
top_performing_schools
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]


#categorize the spendings 
df_school_summary["Spending Ranges (Per Student)"] = pd.cut(budget_per_student,spending_bins,labels = group_names)
df_school_summary
```

```python
##CALCULATIONS

#Average Math Score
avg_math_score_spending = round(df_school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["Average Math Score per School"],2)
#Average Reading Score
avg_reading_score_spending = round(df_school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score per School"],2)
#% Passing Math
students_passing_math_spending = round(df_school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Math per School"],2)
#% Passing Reading
students_passing_reading_spending = round(df_school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading per School"],2)

#Overall Passing Rate (Average of the above two)
overall_pass_rate_spending = (students_passing_math_spending + students_passing_reading_spending)/2



#create a fataframe
scores_by_school_spending = pd.DataFrame({"Average Math Score (Spending)":avg_math_score_spending,
                                     "Average Reading Score (Spending)":avg_reading_score_spending,
                                     "% Passing Math (Spending)":students_passing_math_spending,
                                     "%Passing Reading (Spending)":students_passing_reading_spending ,
                                     "Overall Passing rate (Spending)":overall_pass_rate_spending})

scores_by_school_spending
```
